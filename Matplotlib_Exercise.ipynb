```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Data

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'g')
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("X vs Y Graph")
```

## Exercise 4
** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig , ax = plt.subplots(nrows=1,ncols=2)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y,'g--',lw=3)
ax[0].set_xlim(0,100)
ax[0].set_ylim(0,200)
ax[0].set_xticks([0,20,40,60,80,100])
ax[0].set_yticks([0,50,100,150,200])
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")
ax[0].set_title("X vs Y")

ax[1].plot(x,z,'r',lw=3)
ax[1].set_xlim(0,100)
ax[1].set_ylim(0,10000)
ax[1].set_xticks([0,20,40,60,80,100])
ax[1].set_yticks([0,2000,4000,6000,8000,10000])
ax[1].set_xlabel("X")
ax[1].set_ylabel("Z")
ax[1].set_title("X vs Z")

plt.tight_layout()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,3))
ax[0].plot(x,y,'g--',lw=5)
ax[0].set_xlim(0,100)
ax[0].set_ylim(0,200)
ax[0].set_xticks([0,20,40,60,80,100])
ax[0].set_yticks([0,50,100,150,200])
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")
ax[0].set_title("X vs Y")

ax[1].plot(x,z,'r',lw=5)
ax[1].set_xlim(0,100)
ax[1].set_ylim(0,10000)
ax[1].set_xticks([0,20,40,60,80,100])
ax[1].set_yticks([0,2000,4000,6000,8000,10000])
ax[1].set_xlabel("X")
ax[1].set_ylabel("Z")
ax[1].set_title("X vs Z")

plt.tight_layout()
```
