** Create a function that grabs the email website domain from a string in the form: **

user@domain.com
So for example, passing "user@domain.com" would return: domain.com

```python
def find_domain(email_id):
    return email_id.split("@")[1]
```

```python
find_domain("sushil@gmailcom")
```

```python
find_domain('1234@yahoo.com')
```

** Create a function that counts the number of times the word "dog" occurs in a string. Again ignore edge cases. **

```python
def count_dog(d):
    c=0
    for s in d.lower().split():
        if s=="dog":
            c=c+1
    print(c)
        
```

```python
count_dog("My Dog Name is Revha, DOG dOG doG")
```

```python
count_dog('This dog runs faster than the other dog dude!')
```

**You are driving a little too fast, and a police officer stops you. Write a function to return one of 3 possible results: "No ticket", "Small ticket", or "Big Ticket". If your speed is 60 or less, the result is "No Ticket". If speed is between 61 and 80 inclusive, the result is "Small Ticket". If speed is 81 or more, the result is "Big Ticket". Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday, your speed can be 5 higher in all cases. **

```python
def caught_speeding(speed,birthday):
    if birthday==True:
        speed1=speed-5
    else:
        speed1=speed
    if speed1<=60:
        return "No Ticket!"
    elif speed1<=80:
        return "Small Ticket!"
    else:
        return "Big Ticket!"
        
    
```

```python
caught_speeding(81,True)
```

```python
caught_speeding(81,False)
```

```python
caught_speeding(56,False)
```

```python
caught_speeding(75,False)
```

Q2
WAP to check whether a number is palindrome or not on following
conditions.
1. take input number
2. add number and its reverse
3. check that number is palindrome or not,if not then sum and its reverse

```python
n1=input("Enter the Number :")
n2=int(n1[::-1])
n3=str(int(n1)+n2)
while n3!=n3[::-1]:
    print("Number is :",n3)
    print("Reverse is :",n3[::-1])
    n3=str(int(n3)+int(n3[::-1]))
    print("Number is :",n3)
    print("Reverse is :",n3[::-1])
print("Palindrome is :",n3)
```
